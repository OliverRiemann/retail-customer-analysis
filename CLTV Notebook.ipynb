
# CLTV Notebook

This notebook takes an RFM DataFrame and builds a CLTV DataFrame

```python
from pyspark.sql import DataFrame
from pyspark.sql.functions import col
```

```python
rfm_data = spark.read.format("delta").load("/Volumes/portfolio/cltv_schema/portfolio_crm/rfm_data")
```

```python
def cltv(rfm: DataFrame, retention_rate: float = 0.7) -> DataFrame:
    return rfm.withColumn("CLTV", col("Monetary") * col("Frequency") * retention_rate)
```

```python
cltv = cltv(rfm_data)
cltv.write.format("delta").mode("overwrite").save("/Volumes/portfolio/cltv_schema/portfolio_crm/cltv_data")
```
